```python
from sage.algebras.flag_algebras import * 
```

```python
TG = ThreeGraphTheory
# k4 and k4m (the two induced structures with size 4 excluded)
k4 = TG(4, edges=[[0, 1, 2], [0, 1, 3], [0, 2, 3], [1, 2, 3]])
k4m = TG(4, edges=[[0, 1, 2], [0, 1, 3], [0, 2, 3]])
TG.exclude([k4,k4m])
# 6-vertex design
f=TG(6, edges=[ [1,2,3], [2,3,4], [3,4,5], [4,5,1], [5,1,2], 
               [1,3,0], [2,4,0], [3,5,0], [4,1,0], [5,2,0] ])
degree=TG(3, edges=[ [0,1,2] ], ftype=[0])
p2f4 = TG.generate_flags(4, TG(2, ftype=[0, 1]))
degree_difference = p2f4[2]-p2f4[3]+p2f4[5]-p2f4[6]
print(degree_difference)
```

```python
max_F = TG.optimize_problem(f, 7, maximize=False, positives=[degree-2/7, degree_difference, -degree_difference])
# TG.optimize_problem(f, 6, maximize=True, positives=[degree-0.285714285])
```
